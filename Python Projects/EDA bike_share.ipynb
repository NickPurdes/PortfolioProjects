# ДОСЛІДЖЕННЯ КОРЕЛЯЦІЙ

```python
print(df['temp'].corr(df['hum']))
```

```python
немає кореляції між темп-ю і вологістю
```

```python
Кореляції дадуть нам можливість краще побачити напрям змін між двома факторами протягом певного часу, що може стати 
```

```python
основою для подольшого аналізу даних: виявлення можливих залежностей, наприклад.
```

```python
thenames=['hr','temp','windspeed', 'count']
cor_matrix = df[thenames].corr()
print(cor_matrix)
```

```python
plt.figure(figsize=(14,10))
corr = df[thenames].corr()
sns.heatmap(corr, annot=True,cmap='coolwarm',
        fmt=".3f", xticklabels=thenames,
        yticklabels=thenames)
plt.show()
```

```python
як бачимо, між вказаними змінними і поїздками немає сильних кореляцій , але досить високі показники для годин і температури блтзько 0,4
```

```python
дають нам натяк, що при більш глибокому вивченні цих показників ми все таки можемо знайти певні залежності 
```

```python
# Створення зведеної таблиці
df_hm =df.pivot_table(index = 'hr',columns ='weekday',values ='count')
# Малюємо теплову карту
plt.figure(figsize = (18,7)) 
sns.heatmap(df_hm,  fmt="d", cmap='coolwarm',linewidths=.5, vmin = 0)
plt.show()
```

```python
Бачимо значну к-сть поїздок в будні дні о 8, з 17 по 18 години, у вихідні - з 11 по 17 годину.
```

```python
Щоб збільшити прибуток, на часи пік можна спробувати обережно підвищити ціни, а в час низької активності - запропонувати знижки.
```

# Резюме.

 #### Під час дослідження даних сервісу оренди велосипедів були виявлені наступні факти.

- К-сть поїздок росла з часом, що може свідчити про успішність ведення бізнесу
- К-сть поїздок зареєстрованих споживачів майже у 5 разі вища, ніж незареєстрованих (випадкових)
- Зареєстровані і випадкові споживачі користуються послугами у різний час. Очевидно, що це 2 різних сегментів ринку і різні цільові аудиторії
  які потребують різних методів просування послуги.
      Зареєстровані споживачі, ймовірно, використовують велосипеди для поїздок на роботу і з роботи. Це місцеві клієнти, які мають зручний 
доступ до інфраструктури компанії.
      Випадкові споживачі складають менший сегмент ринку. Необхідні додаткові дослідження для виявлення цього сегменту і цільових аудиторій. Це можуть бути або туристи вихідного дня, або клієнти, які розгядають поїздки велосипедом як дозвілля, оскільки більша частина поїздок припадає на вихідні дні.
- Максимальна к-сть велопарку не повинна бути більшою, ніж 1000 велосипедів. Враховуючи певну сезонність бізнесу, компанія може 
планувати закупки, ремонт і лізинг велопарку відповідним чином
- Найсуттєвішим фактором, який впливає на к-сть поїздок, є час доби. Найбільшою є к-сть поїздок у будні дні з 7 по 9 і з 17 по 18 години, а у вихідні дні- з 11 по 17 годину. Щоб збільшити прибуток, у час пік можна спробувати обережно підвищити ціни, а у час низької активності споживачів запропонувати знижки
- Нічні поїздки не є популярними, але значне стандартне відхилення може свідчити про певний потенціал росту к-сті таких поїздок
